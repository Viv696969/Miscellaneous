```python
data_dir='data'
```

```python
dataset=keras.preprocessing.image_dataset_from_directory(
    data_dir,
    batch_size=32,
    image_size=(256, 256),
    shuffle=True,
    )
```

```python
class_names=dataset.class_names
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
for image,labels in dataset.take(1):
    print(image.shape,labels)
    for i in range(20):
        plt.subplot(4,5,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.axis('off')
        plt.title(f'{class_names[labels[i]],labels[i].numpy()}')
```

```python
def train_val_test_split(dataset,train=0.85,test=0.15,shuffle=True):
    assert(train+test)==1
    ds_size=len(dataset)
    train_size=int(ds_size*train)
    test_size=int(ds_size*test)
    if shuffle:
        dataset.shuffle(5,seed=5)
    
    train_ds=dataset.take(train_size)
    test_ds=dataset.skip(train_size)
#     test_ds=dataset.skip(train_size+val_size)
    print(len(test_ds))
    print(len(dataset))
    print(len(train_ds))
    return train_ds,test_ds
```

```python
def normalize_image(image, label):
    image=tf.image.resize(image, [256,256]) 
    return tf.cast(image/ 255., tf.float32), label
```

```python
ds = dataset.map(normalize_image)
```

```python
train_ds,test_ds=train_val_test_split(ds)
```

```python
for i,j in train_ds.take(10):
    print(i[0].shape)
```

```python
model1=keras.Sequential()
```

```python
# model.add(keras.layers.Conv2D(64,(3,3),activation='relu',input_shape=(32,256,256,3)))
```

```python
model1.add(keras.layers.Resizing(256,256))
model1.add(keras.layers.Conv2D(64,(3,3),activation='relu',input_shape=(32,256,256,3)))
model1.add(keras.layers.MaxPool2D())
model1.add(keras.layers.Conv2D(64,(3,3),activation='relu'))
model1.add(keras.layers.MaxPool2D())
model1.add(keras.layers.Conv2D(32,(3,3),activation='relu'))
model1.add(keras.layers.MaxPool2D())
model1.add(keras.layers.Flatten())
model1.add(keras.layers.Dense(64,activation='relu'))
model1.add(keras.layers.Dense(2,activation='softmax'))
model1.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model1.build(input_shape=(32,256,256,3))
```

```python
model=keras.Sequential()
model.add(keras.layers.Conv2D(64,(3,3),activation='relu',input_shape=(256,256,3)))
model.add(keras.layers.MaxPool2D())
model.add(keras.layers.Conv2D(32,(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D())
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(64,activation='relu'))
model.add(keras.layers.Dense(2,activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
# model.build(input_shape=(32,256,256,3))
```

```python
h1=model.fit(train_ds,validation_data=test_ds,epochs=10)
```

```python
h2=model.fit(train_ds,validation_data=test_ds,epochs=10)
```

```python
model.evaluate(test_ds)
```

```python
def give_label(arr):
#     print(arr)
#     print(np.argmax(arr[0]))
    return class_names[np.argmax(arr[0])]

def predict(img_path):
    img=cv.imread(img_path)
    img=cv.resize(img,(256,256))
    img=img/255
    y_pre=model.predict(
        np.expand_dims(img,0)
        
        )
    return give_label(y_pre)
```

```python
val_dict={'acne':0,'eczema':0}


for img in os.listdir('data/Eczema'):
    img_path=os.path.join('data/Eczema',img)
    label=predict(img_path)
    if label=='Acne':
        val_dict['acne']+=1
    else:
        val_dict['eczema']+=1

```
